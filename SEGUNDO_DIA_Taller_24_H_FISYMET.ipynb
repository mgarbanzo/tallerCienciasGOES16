# Segundo día del taller para las 24 H de FISYMET

Continuamos con señales y Fourier

```
import numpy as np
import pylab as plt
```

```
T1 = 5
T2 = 7
```

```
dt = 0.01
```

```
t1 = np.arange(0,100,dt)
```

```
print(t1)
```

```
F = 1 / dt
print(F)
```

```
df = 1 / 100
print(df)
```

```
y1 = np.sin((2*np.pi*t1)/T1)
```

```
y2 = 0.5 * np.exp(-1j*(2*np.pi*t1)/T1) + 0.5 * np.exp(-1j*(2*np.pi*t1)/T2)
```

```
plt.figure(figsize=(15,5))
plt.plot(t1,y1)
plt.figure(figsize=(15,5))
plt.plot(t1,y2)
```

# Fourier Básico

```
FT1 = np.fft.fft(y1)
```

```
type(FT1), type(FT1[0]), print(FT1[0])
```

```
plt.figure(figsize=(25,5))
plt.plot(np.real(FT1),'-r')
plt.plot(np.imag(FT1),'-b')
plt.plot(np.abs(FT1),'-k')
```

```
plt.figure(figsize=(25,5))
plt.plot(np.angle(FT1),'-b')
```

# Mejorar la vis

```
XFT = np.fft.fftfreq(len(FT1),d=dt)
```

```
plt.figure(figsize=(25,5))
plt.plot(XFT,np.abs(FT1),'-k')
```

```
[np.abs(XFT)<1]
```

```
plt.figure(figsize=(25,5))
plt.plot(XFT[np.abs(XFT)<1],np.abs(FT1[np.abs(XFT)<1]),'-k') # Limitamos los valores del eje X a +-1 Hz
```

# Ahora para y2

```
FT2 = np.fft.fft(y2)
```

```
plt.figure(figsize=(25,5))
plt.plot(XFT,np.abs(FT2),'-k')
```

```
plt.figure(figsize=(25,5))
plt.plot(XFT[np.abs(XFT)<1],np.abs(FT2[np.abs(XFT)<1]),'-k') # Limitamos los valores del eje X a +-1 Hz
```

# Filtro simple

```
W = np.hamming(len(y2))
```

```
plt.figure(figsize=(25,5))
plt.plot(W)
```

```
plt.figure(figsize=(25,5))
plt.plot(y2,'--r')
plt.plot(W,'--k')
plt.plot(W*y2)
```

```
FT2 = np.fft.fft(W * y2)
```

```
plt.figure(figsize=(25,5))
plt.plot(XFT[np.abs(XFT)<1],np.abs(FT2[np.abs(XFT)<1]),'-k')
```

# Incremento de puntos en espectro

```
FT2 = np.fft.fft(W * y2, n=10*len(y2))
```

```
XFT = np.fft.fftfreq(len(FT2),d=dt)
```

```
plt.figure(figsize=(25,5))
plt.plot(XFT[np.abs(XFT)<1],np.abs(FT2[np.abs(XFT)<1]),'-k')
```
